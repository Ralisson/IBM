# Redimensionar todas as imagens

```python
# Imports
import cv2
import glob
import os
import numpy as np 
from pathlib import Path
import matplotlib.pyplot as plt  
```

### Codigo de Exemplo

```python
# Imagem original
image1 = cv2.imread("Imagens/treino/lagarta/original/anticarsia_gemmatalis101_200x200.jpg") 

# Imagem usando interporlação INTER_AREA
image2 = cv2.resize(image1, (128, 128), interpolation = cv2.INTER_AREA) 
  
# Imagem usando interporlação INTER_LINEAR
image3 = cv2.resize(image1, (200, 200), interpolation = cv2.INTER_LINEAR) 
  
# Plot
plt.title("Original") 
plt.imshow(image1) 
plt.show() 

plt.title("Resize 1") 
plt.imshow(image2) 
plt.show() 

plt.title("Resize 2") 
plt.imshow(image3) 
plt.show() 
```

### Aplicar para todo o diretorio 'lagarta'

```python
inputFolder = 'Imagens/treino/lagarta/original'
newFolder = 'Imagens/treino/lagarta/resized'

# Criamos o PATH (caminho)
dir_path = Path(newFolder)

# Verificamos se o diretório já existe e se não existir, criamos
if dir_path.exists():
    print('O diretório já existe.')
else:
    os.mkdir(dir_path)
    
i = 0

for img in glob.glob(inputFolder + "/*.*"):
    image = cv2.imread(img)
    
    # Imagem usando interporlação INTER_AREA
    #imgResized = cv2.resize(image, (96,96), interpolation = cv2.INTER_AREA) 
    
    # Imagem usando interporlação INTER_LINEAR
    imgResized = cv2.resize(image, (200, 200), interpolation = cv2.INTER_LINEAR) 

    cv2.imwrite(newFolder + "/image%04i.jpg" %i, imgResized)
    i += 1
    
cv2.destroyAllWindows()    
```
