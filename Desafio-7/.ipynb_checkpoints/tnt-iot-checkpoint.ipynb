```python
# Funcao para carregar os dados do banco de dados NoSQL Cloudant
def obter_dados(user, password, account, db_name):

    client = Cloudant(user, password, account=account, connect=True, auto_renew=True)

    db = client[db_name]

    response = db.all_docs(limit=20000, include_docs=True)

    docs = []
    for r in response["rows"]:
        docs.append(r['doc'])
        
    # Cria um data frame:
    df = pd.DataFrame(data=docs)

    # Elimina as colunas desnecessárias:
    df.drop(labels=['_id', '_rev', 'row'], axis=1, inplace=True)
    
    # Elimina as duplicidades:
    linhas_duplicadas = df.duplicated().sum()
    if (linhas_duplicadas > 0):
        df.drop_duplicates(subset=None, keep='first', inplace=True)

    # Retorna o dataframe criado:
    return df
```

```python
# ***** Função para salvar (save) um objeto python qualquer em um arquivo "pickle":
#
def pickle_object_save(path=".", file="None", object_name=None, msg=None):
    '''
    Input:
        "path": diretório (path) onde o arquivo será criado.
        "file": nome do arquivo que será armazenado (salvo) o objeto python.
        "object_name": nome do objeto python que será armazenado (salvo) no arquivo.
        "msg": mensagem que será impressa na tela (default é None, ou seja,
               não será impressa uma mensagem).

    Output: None
    '''
    # Código da função:

    # Prepara o nome completo do arquivo piclke que será criado:
    object_file = os.path.join(path, file)
    
    try:
        # Abre o arquivo para o modo escrita (write):
        pickle_out = open(object_file,"wb")
        
        # Faz o 'dump' do objeto ("object_name") e salva os dados no arquivo ("file"):
        pickle.dump(object_name, pickle_out)

        # Fecha o arquivo "pickle":
        pickle_out.close()

        # Verifica se existe uma mensagem para ser impressa na tela:
        if (msg != None):
            # Imprime na tela a mensagem:
            print(msg)

    except FileNotFoundError as error:
        # Erro encontrado na abertura do arquivo:
        print(error)

    except:
        # Erro encontrado ao fazer o "dump" do objeto:
        print("I can not save the '{}' object".format(object_name))
        
        # Fecha o arquivo "pickle":
        pickle_out.close()
####################################################################################################
# ***** Função para carregar (load) um objeto python qualquer, armazenado em um arquivo "pickle":
#
def pickle_object_load (path=".", file="None", msg=None):
    '''
    Input:
        "path": diretório (path) do arquivo que será carregado.
        "file": nome do arquivo que será carregado.
        "msg": mensagem que será impressa na tela (default é None, ou seja,
               não será impressa uma mensagem).

    Output:
        "obj": retorna os dados do "objeto" armazenado no arquivo.
    '''
    # Código da função:

    # Prepara o nome completo do arquivo piclke que será carregado (load):
    object_file = os.path.join(path, file)
    
    try:
        # Abre o arquivo para o modo leitura (read):
        pickle_in = open(object_file,"rb")
        
        # Faz a leitura do arquivo e carrega os dados no objeto ("obj"):
        obj = pickle.load(pickle_in)

        # Fecha o arquivo "pickle":
        pickle_in.close()

        # Verifica se existe uma mensagem para ser impressa na tela:
        if (msg != None):
            # Imprime na tela a mensagem:
            print(msg)

        # Retorna os dados carregados:
        return obj
    
    except FileNotFoundError as error:
        # Erro encontrado na abertura do arquivo:
        print(error)
        # Retorna um valor nulo (None):
        return None

    except ValueError:
        # Erro encontrado na leitura do arquivo:
        print("I can not upload the '{}' file".format(file))
        
        # Fecha o arquivo "pickle":
        pickle_in.close()
        
        # Retorna um valor nulo (None):
        return None
```

### `5-` Analisando e removendo as datas inválidas:

```python
# Função para obter as datas inválidas:
def obter_datas_invalidas(df, dt):
    contagem = 0
    df_data_inv = pd.DataFrame()

    for data in dt:
        index = (df.loc[:, 'Tempo'] == data)
        contagem += df[index].shape[0]
        df_data_inv = pd.concat([df_data_inv, df[index]], ignore_index=True)
    
    return (contagem, df_data_inv)
```

```python
# Obtendo as datas inválidas:
datas_inv = ['2016-2-30', '2017-2-30', '2018-2-30', '2019-2-30','2020-2-30']
contagem, df_data_inv = obter_datas_invalidas(df=df_data, dt=datas_inv)
print("Contagem de linhas com datas inválidas: {}".format(contagem))
df_data_inv.head()
```

```python
# Função para remover as datas inválidas:
def remover_datas_invalidas(df, dt):
    contagem = 0
    df_copy = df.copy() # Faz uma cópia do data frame.

    for data in dt:
        flag = list((df_copy.loc[:, 'Tempo'] == data))
        rows_inv = list(df_copy[flag].index)
        contagem += len(rows_inv)
        df_copy.drop(index=rows_inv, inplace=True) # Remove as linhas com datas inválidas.
    
    return (contagem, df_copy.reset_index(drop=True))
```

```python
# Removendo as datas inválidas:
contagem, df_data = remover_datas_invalidas(df=df_data, dt=datas_inv)
print("Contagem de linhas removidas com datas inválidas: {}".format(contagem))
print(df_data.shape)
df_data.head()
```

```python
# Verificando se as datas inválidas foram removidas:
datas_inv = ['2016-2-30', '2017-2-30', '2018-2-30', '2019-2-30','2020-2-30']
contagem, df_data_inv = obter_datas_invalidas(df=df_data, dt=datas_inv)
print("Contagem de linhas com datas inválidas: {}".format(contagem))
df_data_inv.head()
```

**Podemos observar que as datas inválidas foram removidas do dataset!**
